```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
import psycopg2
```

```python
poki520_csv = 'templates/data/pokedex_0520.csv'
pokimain_df = pd.read_csv(poki520_csv, encoding='UTF-8')
pokimain_df.head()
```

```python
poki420_csv = 'templates/data/pokedex_0420.csv'
pokilegends_df = pd.read_csv(poki420_csv, encoding='UTF-8')
pokilegends_df.head()
```

```python
effectiveness_csv = 'templates/data/effectiveness_by_type.csv'
effectiveness_by_type_df = pd.read_csv(effectiveness_csv, encoding='UTF-8')
effectiveness_by_type_df.head()
```

```python
# Listing all columns in main DF  
pokimain_df.columns
```

```python
# Listing all columns in main DF  
pokilegends_df.columns
```

```python
legendary_df = pokilegends_df[["pokedex_number", "name", "generation","is_sub_legendary", "is_legendary", "is_mythical", 'total_points', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]
legendary_df.head()
```

```python
pd.set_option('display.max_rows', None) 
legendary_cleandf = legendary_df.loc[(((legendary_df['is_sub_legendary'] != 0) | (legendary_df['is_legendary'] != 0) | (legendary_df['is_mythical'] != 0)))].dropna(how='all', axis=0)
legendary_cleandf
```

```python
legendary_cleandf['Special'] = 'Legendary'
legendary_cleandf = legendary_cleandf[["pokedex_number", "name", "generation",'total_points', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'Special']]
legendary_cleandf
```

```python
starters = ["Bulbasaur", "Ivysaur", "Venusaur", "Charmander", "Charmeleon", "Charizard", "Squirtle", 'Wartortle', 'Blastoise',
'Chikorita', "Bayleef", "Meganium", "Cyndaquil", "Quilava", "Typhlosion", "Totodile", 'Croconaw', 'Feraligatr',
"Treecko", 'Grovyle', 'Sceptile', "Torchic", "Combusken", "Blaziken", "Mudkip", "Marshtomp", 'Swampert',
"Turtwig", "Grotle", "Torterra", 'Chimchar', 'Monferno', "Infernape", "Piplup", "Prinplup", "Empoleon",
"Snivy", 'Servine', "Serperior", "Tepig", 'Pignite', "Emboar", "Oshawott", "Dewott", "Samurott",
'Chespin', "Quilladin", 'Chesnaught', "Fennekin", "Braixen", "Delphox", 'Froakie', 'Frogadier', "Greninja",
"Rowlet", 'Dartrix', "Decidueye", "Litten", "Torracat", "Incineroar", "Popplio", 'Brionne', "Primarina", 
"Grookey", 'Thwackey', "Rillaboom", "Scorbunny", "Raboot", "Cinderace", "Sobble", "Drizzile", "Inteleon"] 
```

```python
for pokemon in pokimain_df["name"]:
    starter_index = pokimain_df[pokimain_df["name"] == pokemon].index
    if pokemon in starters:
        pokimain_df.loc[starter_index, "is_starter"] = 1
    else: 
        pokimain_df.loc[starter_index, "is_starter"] = 0
```

```python
pokimain_df.head()
```

```python
starter_cleandf = pokimain_df[["pokedex_number", "name", "generation", "is_starter",'total_points', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]
starter_cleandf = starter_cleandf.loc[(starter_cleandf['is_starter'] != 0)]
starter_cleandf
```

```python
starter_cleandf['Special'] = 'Starter'
starter_cleandf = starter_cleandf[["pokedex_number", "name", "generation",'total_points', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'Special']]
starter_cleandf
```

```python
legendarystarters_df = pd.concat([legendary_cleandf, starter_cleandf], axis=0)
legendarystarters_df
```

```python
pokimain_df = pokimain_df[['pokedex_number', 'name',
       'generation', 'status', 'species', 'type_number', 'type_1', 'type_2',
       'height_m', 'weight_kg', 'abilities_number', 'ability_1', 'ability_2',
       'ability_hidden', 'total_points', 'hp', 'attack', 'defense',
       'sp_attack', 'sp_defense', 'speed', 'catch_rate', 'base_friendship',
       'base_experience', 'growth_rate', 'egg_type_number', 'egg_type_1',
       'egg_type_2', 'percentage_male', 'egg_cycles', 'against_normal',
       'against_fire', 'against_water', 'against_electric', 'against_grass',
       'against_ice', 'against_fight', 'against_poison', 'against_ground',
       'against_flying', 'against_psychic', 'against_bug', 'against_rock',
       'against_ghost', 'against_dragon', 'against_dark', 'against_steel',
       'against_fairy', "is_starter"]]
pokimain_df.head()
```

```python
type_df = pokimain_df[["type_1","type_2",'against_normal',
       'against_fire', 'against_water', 'against_electric', 'against_grass',
       'against_ice', 'against_fight', 'against_poison', 'against_ground',
       'against_flying', 'against_psychic', 'against_bug', 'against_rock',
       'against_ghost', 'against_dragon', 'against_dark', 'against_steel',
       'against_fairy']]
type_df.head()
```

```python
effectiveness_df = type_df[type_df['type_2'].isnull()]
effectiveness_df = effectiveness_df.drop_duplicates(subset=['type_1'])
effectiveness_df = effectiveness_df[["type_1",'against_normal',
       'against_fire', 'against_water', 'against_electric', 'against_grass',
       'against_ice', 'against_fight', 'against_poison', 'against_ground',
       'against_flying', 'against_psychic', 'against_bug', 'against_rock',
       'against_ghost', 'against_dragon', 'against_dark', 'against_steel',
       'against_fairy']]
effectiveness_df
```

```python
# Connection to PostgreSQL server
rds_connection_string = "postgres:postgres@localhost:5432/pokemon.db"
engine = create_engine(f"postgresql://{rds_connection_string}")
```

```python
engine.table_names()
```

```python
# Dataframes to tables in PostgresSQL
pokimain_df.to_sql(name='pokemon_data', con=engine, if_exists='append', index=False)
legendarystarters_df.to_sql(name='legendary_and_starters', con=engine, if_exists='append', index=False)
effectiveness_df.to_sql(name='type_data', con=engine, if_exists='append', index=False)
effectiveness_by_type_df.to_sql(name='effectiveness_by_type_data', con=engine, if_exists='append', index=False)
```
